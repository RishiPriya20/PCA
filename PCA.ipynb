**Task 2**

**Dimensionality Reduction with PCA**

```python
from sklearn.preprocessing import StandardScaler

target_column = 'Type'
#Seperateing features and target
X = df.drop(target_column, axis=1)
y = df[target_column]
#Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

from sklearn.preprocessing import LabelEncoder

# Encode the target column if needed
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)
```

```python
#Implementing PCA to reduce the dimensionality of the dataset
from sklearn.decomposition import PCA
#Applyinf PCA
pca = PCA()
X_pca = pca.fit_transform(X_scaled)
```

```python
#Determinine the optimal number of principal components using techniques like scree plot or cumulative
#Scree plot
plt.plot(range(1, X_scaled.shape[1] + 1), pca.explained_variance_ratio_, 'o-')
plt.xlabel('Number of Components')
plt.ylabel('Explained Variance Ratio')
plt.title('Scree Plot')
plt.show()
```

```python
#Cumulative exlained variance
cumulative_variance = pca.explained_variance_ratio_.cumsum()
plt.plot(range(1, X_scaled.shape[1] + 1), cumulative_variance, 'o-')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Cumulative Explained Variance')
plt.show()
```

```python
#Transform the original dataset into the principal components
#Choosing the number of components based on the cumulative explained variance
n_components = 2
pca =PCA(n_components=n_components)
X_pca = pca.fit_transform(X_scaled)
```

**Observation**

1. Standardization: The features were standardized to have zero mean and unit variance.
Used 'StandardScaler' from 'sklearn.preprocessing' to scale the data.
This step checks that all features contribute equally to the PCA process, preventing any feature with a large scale from dominating the analysis
Encoding Target Variable: The Target variable 'Type' was encoded into numerical values to prepare for PCA
2. Used 'LadelEncoder' from 'sklearn.preprocessing' to convert the target labels into numerical form.
This conversion allows the target variable to be used in subsequent ML tasks.
Applying PCA: PCA was applied to reduce the dimensionality of the dataset.
Used 'PCA' from 'sklearn.decomposition' to transform the standardized features into principal components.
3. Determining the Optimal Number of Principal Components:
Screen Plot: Plotted the explained variance ratio for each principal component.
The screen plot helps to visualize how much variance each component explains. Typically there is a point where the explained variance starts to level off indicating a suitable number of components to retain.
4. Cumulative Explained Variance: Plotted the cumulative explained variance to show the total variance explained by the components.
The cumulative variance plot helps a sufficient amount of the total variance.For suppose if the plot reaches around 80-90% cumulative variance with the first tew components those components are generally chosen.
5. Transforming the Data: It  Reduces the dataset to a smaller number of dimensions based on the PCA results.
Chose 2 principal components based on the cumulative explained variance plot and re-applied PCA with 'n-components=2'
The original data was transformed into two principal components simplifying the dataset while retaining most of the variance.



**Task 3**

**Clustering with original Data**

```python
#Clustering with original Data
#Applying a clustering algorithm
from sklearn.cluster import KMeans
#Apply K-means clustering
kmeans = KMeans(n_clusters=3)
clusters = kmeans.fit_predict(X_scaled)
df['Cluster'] = clusters
#Visualize the clustering results using appropriate plots
plt.figure(figsize=(10, 6))
sns.scatterplot(x=X_scaled[:, 0], y=X_scaled[:, 1], hue=clusters, palette='viridis')
plt.title('K-means Clustering on Original Data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
#Evalute the clustering performance using metrics such as silhouette score or Davies-Bouldin index
from sklearn.metrics import silhouette_score, davies_bouldin_score

# Silhouette Score
silhouette_avg = silhouette_score(X_scaled, clusters)
print(f'Silhouette Score: {silhouette_avg}')

# Daviesâ€“Bouldin Index
dbi = davies_bouldin_score(X_scaled, clusters)
print(f'Daviesâ€“Bouldin Index: {dbi}')
```

**Observation**

1. Clustering Algorithm:
To apply K-means clustering to the original standardized dataset.
Used the K-means algorithm with 'n-cluster =3' which means the algorithm tries to find 3 distinct clusters in the data.
The K-means algorithm assigned each data point to one of the 3 clusters and these clusters were added as a new column named 'Cluster' in the dataset.
2. Visualization of Clustering Results:
Scatter plot: Plotted a scatter plot of the first two features and colored the points based on their cluster assignments.
The scatter plot shows how the data points are grouped into clusters. The colors represent different clusters, making it possible to visually assess how well the clusters are separated and if there is any overlap between them.
3. Clustering Performance Evaluation:
Silhouette Score: It calculated the silhouette sore to measure how similar each point is to its cluster compared to other clusters. The score ranges from -1 to 1 where a higher score indicates better-defined clusters.
The silhouette score was '0.2849' which suggests that the cluster have some separation but they are not very distinct.
Davies-Bouldin Index:
Calculated the Davies-Bouldin index to evaluated the average similarity ration of each cluster with its most similar cluster. A lower index indicates better clustering.
The Davies-Bouldin index was '1.3892' indicating that the clusters might not be well-separated as the index value suggests some overlap between clusters.
